# Stage1

    1) Calculate % growth of funding between year 2008 & 2009.

    2) If funding is positive, tag it as 1, if funding is negative tag it as 0. This is the target variable.
    pandas.core.frame.DataFrame

```python
import pandas as pd
pd.options.mode.chained_assignment = None 
```

```python
dataframe=pd.read_csv("2010_Federal_STEM_Education_Inventory_Data_Set.csv")
```

```python
dataframe.head()
```

```python
dataframe=dataframe.apply(pd.to_numeric, errors='ignore') #convert object type to float
dataframe.info() #check column type
```

```python
dataframe['Target']="" #new column initialized
dataframe['Growth of Funding(%)']="" #new column initialized
for itr in range(0,len(dataframe)):
        if dataframe['C1) Funding FY2008'][itr]!=0: #handle /0 case
            percentage=round(((dataframe['C2) Funding FY2009'][itr]-dataframe['C1) Funding FY2008'][itr])/dataframe['C1) Funding FY2008'][itr]),2)*100
        elif dataframe['C1) Funding FY2008'][itr]==0 and dataframe['C2) Funding FY2009'][itr]==0:
            percentage=0
        elif dataframe['C1) Funding FY2008'][itr]==0 and dataframe['C2) Funding FY2009'][itr]!=0: #since the initial funding for that program was 0 then if some funding is done in 2009 it should be considered 100% growth 
            percentage=100
            
        dataframe['Growth of Funding(%)'][itr]=percentage 
        if percentage>0:
            dataframe['Target'][itr]=1
        elif percentage<=0:
            dataframe['Target'][itr]=0
            
```

```python
dataframe.to_csv("Stage1.csv",index=False)
```
