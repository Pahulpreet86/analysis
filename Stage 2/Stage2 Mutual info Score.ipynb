```python
import pandas as pd
from sklearn.metrics.cluster import normalized_mutual_info_score
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
dataframe=pd.read_csv("Stage1.csv")
column=dataframe.columns.unique()
column=column[6:]
```

```python
result=[]
for itr in range(0,len(column)):
    score=normalized_mutual_info_score(dataframe[column[itr]],dataframe["Target"])
    result.append((round(score,2),column[itr]))
```

```python
result #between 0 (no mutual information) and 1 (perfect correlation)
```

```python
final_result=[]
for itr in range(0,len(column)):
    scores=[]
    for itr1 in range(0,len(column)):
        score=normalized_mutual_info_score(dataframe[column[itr]],dataframe[column[itr1]])
        scores.append(round(score,2))
    final_result.append(scores)
```

```python
plt.figure(figsize=(15,10))
ax = sns.heatmap(final_result,xticklabels=column,yticklabels=column)
plt.title("Visualization of Normalized Mutual Information Score")
plt.show()
```
