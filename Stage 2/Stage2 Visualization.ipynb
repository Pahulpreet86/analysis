```python
import pandas as pd
pd.options.mode.chained_assignment = None 
import matplotlib.pyplot as plt  
```

```python
dataframe=pd.read_csv("Agency.csv")
datavisual=dataframe.groupby('Agency').size().reset_index(name='counts')
datavisual=datavisual.sort_values(by=['counts'])
datavisual = datavisual.reset_index(drop=True)
plt.barh(datavisual['Agency'],datavisual['counts'],color='#3877AD')
plt.xlabel("Count")
plt.title("Agency Distribution")
plt.show()
```
